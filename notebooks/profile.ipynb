# Profile
Measurements of progress over the course the algorithm execution

## Generate Dataset
Use `python/profile.py` to run the data collection

## Load Dataset

```python
import matplotlib.pyplot as plt
import pandas as pd

dataset = pd.DataFrame(pd.read_csv("experiments/analysis/profile/compas.csv", delimiter=","))
```

## Convergence

```python
time = dataset['time']
lowerbound = dataset['lowerbound']
upperbound = dataset['upperbound']

fig = plt.figure(figsize=(10, 4), dpi=100)
plt.plot(time, lowerbound, label='lowerbound', markersize=2, marker='o', linewidth=1)
plt.plot(time, upperbound, label='upperbound', markersize=2, marker='o', linewidth=1)
plt.xlabel('Time (s)')
plt.ylabel('Objective Upper and Lower Bounds')
plt.grid()
plt.legend()
plt.title('Objective Bounds vs Time')
```

## Graph Cache Efficiency

```python
time = dataset['time']
size = dataset['graph_size']
hits = dataset['graph_hits']
hit_rate = dataset['graph_hits'] / (dataset['graph_size'] + dataset['graph_hits'])

fig = plt.figure(figsize=(10, 4), dpi=100)
plt.plot(time, size, label='size', markersize=2, marker='o', linewidth=1)
plt.xlabel('Time (s)')
plt.ylabel('Graph Size')
plt.grid()
plt.legend()
plt.title('Graph Size vs Time')

fig = plt.figure(figsize=(10, 4), dpi=100)
plt.plot(time, hit_rate, label='hit rate', markersize=2, marker='o', linewidth=1)
plt.xlabel('Time (s)')
plt.ylabel('Graph Hit Rate')
plt.grid()
plt.legend()
plt.title('Graph Hit Rate vs Time')
```

## Load Distribution

```python
queue_partitions = 10

time = dataset['time']
size = dataset['queue_size']

fig = plt.figure(figsize=(10, 4), dpi=100)
plt.plot(time, size, label='global queue', markersize=2, marker='o', linewidth=1)
plt.ylabel('Global Queue Size')
plt.xlabel('Time (s)')
plt.grid()
plt.legend()
plt.title('Global Queue Size vs Time')

fig = plt.figure(figsize=(10, 4), dpi=100)
for k in range(queue_partitions):
    plt.plot(time, dataset['local_queue_{}_size'.format(k)], label='local queue {}'.format(k), markersize=2, marker='o', linewidth=1)
plt.ylabel('Local Queue Size')
plt.xlabel('Time (s)')
plt.grid()
plt.legend()
plt.title('Local Queue Size vs Time')

```
